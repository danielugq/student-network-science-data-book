### 2.e)

```python
import time
time_dict = {}

for file in files:
    for _ in range(10):
        time_start = time.time()
        G = nx.read_gml(file)
        D = nx.to_numpy_array(G, weight="cost")
        tsp = TSP()
        tsp.read_mat(D)
        two_opt = TwoOpt_solver(initial_tour='NN', iter_num=10000)
        two_opt_tour = tsp.get_approx_solution(two_opt)
        time_end = time.time()
        if _ == 0:
            time_dict[file] = [time_end - time_start]
        else:
            time_dict[file].append(time_end - time_start)
time_dict
```

```python
n_nodes = [52, 58, 180, 229, 439]
avg_times = [np.mean(times) for times in time_dict.values()]
plt.scatter(n_nodes, avg_times)
plt.xlabel("Number of nodes")
plt.ylabel("Average time")
plt.title("Average time of 2-opt approximate solution for the Traveling Salesman Problem")
plt.show()
```
